```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
import pandas as pd
import numpy as np
df = pd.DataFrame(data.data, columns = data.feature_names)
y = pd.DataFrame(data.target.reshape(-1,1), columns = ['Cancer'])
df = pd.concat([df, y], axis = 1)


import tkinter as tk 
import statsmodels.formula.api as smf
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

#train_features = df.columns.difference(['Cancer'])
X = df[['mean radius','mean texture', 'radius error', 'worst smoothness', 'worst symmetry']]
Y = df['Cancer']

# with sklearn
regr = LogisticRegression()
regr.fit(X, Y)


# with statsmodels
#X = sm.add_constant(X) # adding a constant

#model = smf.glm(formula='Cancer ~ ' + "+".join(train_features), data=train_X, family=sm.families.Binomial()).fit()
#predictions = model.predict(X) 

# tkinter GUI
root= tk.Tk() 
root.title('Breast Cancer Classification')
canvas1 = tk.Canvas(root, width = 500, height = 300)
canvas1.pack()

# with sklearn
Intercept_result = ('Intercept: ', regr.intercept_)
label_Intercept = tk.Label(root, text=Intercept_result, justify = 'center')
canvas1.create_window(260, 220, window=label_Intercept)

# with sklearn
Coefficients_result  = ('Coefficients: ', regr.coef_)
label_Coefficients = tk.Label(root, text=Coefficients_result, justify = 'center')
canvas1.create_window(260, 240, window=label_Coefficients)

# meanradius label and input box
label1 = tk.Label(root, text='meanradius: ')
canvas1.create_window(100, 100, window=label1)

entry1 = tk.Entry (root) # create 1st entry box
canvas1.create_window(270, 100, window=entry1)

# meantexture label and input box
label2 = tk.Label(root, text=' meantexture: ')
canvas1.create_window(100, 120, window=label2)

entry2 = tk.Entry (root) # create 2nd entry box
canvas1.create_window(270, 120, window=entry2)

# radiuserror label and input box
label3 = tk.Label(root, text=' radiuserror: ')
canvas1.create_window(100, 140, window=label3)

entry3 = tk.Entry (root) # create 3rd entry box
canvas1.create_window(270, 140, window=entry3)

# worstsmoothness label and input box
label4 = tk.Label(root, text=' worstsmoothness: ')
canvas1.create_window(100, 160, window=label4)

entry4 = tk.Entry (root) # create 4rth entry box
canvas1.create_window(270, 160, window=entry4)

# worstsymmetry label and input box
label5 = tk.Label(root, text=' worstsymmetry: ')
canvas1.create_window(100, 180, window=label5)

entry5 = tk.Entry (root) # create 5th entry box
canvas1.create_window(270, 180, window=entry5)



def values(): 
    global meanradius #our 1st input variable
    meanradius = float(entry1.get()) 
    
    global meantexture #our 2nd input variable
    meantexture = float(entry2.get()) 
    
    global radiuserror #our 1st input variable
    radiuserror = float(entry3.get()) 
    
    global worstsmoothness #our 2nd input variable
    worstsmoothness = float(entry4.get()) 
    
    global worstsymmetry #our 2nd input variable
    worstsymmetry = float(entry5.get()) 
    
    Prediction_result  = ('Predicted Breast Cancer: ', regr.predict_proba([[meanradius ,meantexture, radiuserror, worstsmoothness, worstsymmetry]]))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(260, 200, window=label_Prediction)
    
button1 = tk.Button (root, text='Predict Breast Cancer',command=values, bg='orange') # button to call the 'values' command above 
canvas1.create_window(270, 280, window=button1)


#plot 1st barplot 


figure3 = plt.Figure(figsize=(5,4), dpi=100)
ax3 = figure3.add_subplot(111)
sns.barplot(df['Cancer'].astype(float),df['mean radius'].astype(float), color = 'r', ax = ax3)
barplot3 = FigureCanvasTkAgg(figure3, root) 
barplot3.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax3.legend() 
ax3.set_ylabel('mean radius')
ax3.set_xlabel('Cancer Category')
ax3.set_title('mean radius Vs. Cancer')

#plot 2nd barplot 


figure4 = plt.Figure(figsize=(5,4), dpi=100)
ax4 = figure4.add_subplot(111)
sns.barplot(df['Cancer'].astype(float),df['radius error'].astype(float), color = 'g', ax = ax4)
barplot3 = FigureCanvasTkAgg(figure4, root) 
barplot3.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax4.legend() 
ax4.set_ylabel('radius error')
ax4.set_xlabel('Cancer Category')
ax4.set_title('radius error Vs. Cancer')

#root.configure(background='black')
root.mainloop()
```
